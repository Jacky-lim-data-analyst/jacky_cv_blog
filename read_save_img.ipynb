# Setup

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

if not cv.useOptimized():
    cv.setUseOptimized(True)

def display_inline_img(arr: np.ndarray, title=None, xyticks=False, grayscale=True):
    if grayscale:
        plt.imshow(arr, cmap = plt.cm.gray, vmin=0, vmax=255)
    else:
        plt.imshow(arr)
    plt.title(title)
    if not xyticks:
        plt.xticks([]), plt.yticks([])
    plt.show()
```

# Read image
There are a variety of image file extensions, such as TIFF(.tif), GIF(.gif), JPEG(.jpg, .jpeg), PNG(.png) and WebP(.webp) and etc.
We can load the image in grayscale or color image by specifying the flag argument of `cv.imread()` function.

```python
img = cv.imread("images/flower.jpeg", 1)
img_grayscale = cv.imread("images/flower.jpeg", 0)
```

```python
# grayscale
display_inline_img(img_grayscale, title="flower")
```

When we use `plt.imshow()` function to display the image, it expects the image to be in RGB format. So, we need to swap the order of the blue and red channels since `cv.imread()` outputs BGR image.

```python
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
display_inline_img(img_rgb, title="flower", grayscale=False)
```

We can also display images with `cv.imshow()`. Press Esc key to close the pop-up window.

```python
cv.imshow("flower", img)
cv.waitKey(0)
cv.destroyAllWindows()
```

The png image is distinct from other image extension: it supports extra channel, called $\alpha$ channel, which controls transparency.

```python
img = cv.imread("images/flower.jpeg", cv.IMREAD_UNCHANGED)
img_png = cv.imread("images/watch_star.png", cv.IMREAD_UNCHANGED)
print(f"The number of channels of jpeg image: {img.shape[-1]}")
print(f"The number of channels of png image: {img_png.shape[-1]}")
```

# Save image
We can save image in different file extension.

```python
img_wp = cv.imread("images/chicken.webp")
cv.imwrite("images/chicken.jpeg", img_wp)
```
